# Exclude Thumbs.db and sort image files in the correct frame order

```python
from pathlib import Path
import re

img_dir = Path("resimler")

def frame_number(p: Path) -> int:
    # Extract the first number from the filename (e.g., "resim94" -> 94)
    m = re.search(r"(\d+)", p.stem)
    return int(m.group(1)) if m else 10**9

# Collect only image files (ignore Thumbs.db and anything else)
files = list(img_dir.glob("*.JPG")) + list(img_dir.glob("*.jpg")) + list(img_dir.glob("*.JPEG")) + list(img_dir.glob("*.jpeg"))

# Sort numerically by frame number
files_sorted = sorted(files, key=frame_number)

print("Total image files:", len(files_sorted))
print("First 10 (sorted):", [p.name for p in files_sorted[:10]])
print("Last 10 (sorted):", [p.name for p in files_sorted[-10:]])
```

```python
from PIL import Image

# 1) Load frames in the sorted order we already created (files_sorted)
frames = [Image.open(p).convert("RGB") for p in files_sorted]

# 2) Make sure all frames have the same size (safe guard)
base_size = frames[0].size
frames = [im.resize(base_size) for im in frames]

# 3) Export as GIF
output_path = "animation.gif"
duration_ms = 80  # speed: milliseconds per frame

frames[0].save(
    output_path,
    save_all=True,
    append_images=frames[1:],
    duration=duration_ms,
    loop=0,
    optimize=True
)

print(f"GIF created: {output_path} | Frames: {len(frames)} | Duration per frame: {duration_ms} ms")
```
